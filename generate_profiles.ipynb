```python
import os
import json
from tqdm import tqdm
from pathlib import Path
from classify_questions import convert_persona_to_text

PROJECT_ROOT_PATH = "sim"
project_root = Path(PROJECT_ROOT_PATH)

persona_json_dir = project_root / "data" / "mega_persona_json" / "mega_persona"
output_dir = project_root / "text_simulation"

try:
    json_files = [f for f in os.listdir(persona_json_dir) 
                  if f.endswith('.json') and f.startswith('pid_')]
    
    print(f"Converting {len(json_files)} personas...")

    with open("question_id_mapping.json", "r") as f:
        question_mappings = json.load(f)

    for json_file in tqdm(json_files, desc="Converting personas"):
        input_path = persona_json_dir / json_file

        for category in question_mappings:
            topic = category["Topic"]
            tasks = category["Tasks"]

            output_dir = output_dir / json_file.replace(".json", "")
            output_dir.mkdir(parents=True, exist_ok=True)

            for task in tasks:
                task_name = task["Task"]
                question_ids = task["QuestionIDs"]
                output_path = output_dir / f"{task_name}.txt"
                convert_persona_to_text(str(input_path), str(output_path), topic, question_ids)
                        
    print(f"\n✅ Conversion complete.")
    
except Exception as e:
    print(f"❌ Error converting personas: {e}")
```
